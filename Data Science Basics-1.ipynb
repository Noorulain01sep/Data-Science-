**Q # 2:** NumPy (Numerical Python) is used for Working with arrays (like tables of numbers). Consider an array of any size and values to practice following functions.

```python
import numpy as np

arr_2D = np.array([[1,2,],[2,5]])
arr_1D = np.arange(1,20,4)
arr2_1D = np.arange(10)
arr2_2D = np.arange(1, 5).reshape(2, 2)

print("2D array:\n", arr_2D)
print("1D array using arrange start, stop & step: ", arr_1D)
print("1D array using arrange:", arr2_1D)
print("2D array using arrange & reshape:\n", arr2_2D)

print("2D array with zeros: \n", np.zeros((2,2),dtype=int))
print("2D array with ones: \n", np.ones((3,2)))

print("Shape of an array: ", arr2_2D.shape)
print("Size of an array: ", arr2_2D.size)
print("Data type of an array: ", arr2_2D.dtype)

print("Mean of all elements in array: " , np.mean(arr_1D))
print("Mean along rows:", np.mean(arr_2D, axis=1))
print("Mean along cols:", np.mean(arr_2D, axis=0))

print("Median of array: " , np.median(arr_1D))

print("Standard Deviation of array: " , np.std(arr_1D))

dot_product_result=np.dot(arr_2D,arr2_2D)
print("Dot product of two array are: ",dot_product_result)
```
